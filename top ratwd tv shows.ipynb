```
# ============================================
# EXPLORATORY DATA ANALYSIS (EDA)
# DATASET: TOP RATED TV SHOWS
# ============================================

# -------- 1. IMPORT LIBRARIES --------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------- 2. LOAD DATASET --------
df = pd.read_csv("top_rated_tv.csv")

print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# -------- 3. DATA CLEANING --------
print("\nMissing Values:")
print(df.isnull().sum())

# Drop missing values (simple approach)
df = df.dropna()

print("\nAfter cleaning, dataset shape:", df.shape)

# -------- 4. DATA PREPROCESSING --------
# Convert date column if exists
if 'first_air_date' in df.columns:
    df['first_air_date'] = pd.to_datetime(df['first_air_date'], errors='coerce')

# Feature selection (adjust if needed)
features = ['popularity', 'vote_average', 'vote_count']
df = df[features]

# -------- 5. DATA VISUALIZATION --------

# GRAPH 1: Distribution of Ratings
plt.figure()
sns.histplot(df['vote_average'], bins=20)
plt.title("Distribution of Vote Average")
plt.xlabel("Vote Average")
plt.ylabel("Frequency")
plt.show()

print("\nObservation:")
print("Most TV shows have ratings clustered between mid to high values.")

# -----------------------------------

# GRAPH 2: Popularity vs Vote Average
plt.figure()
sns.scatterplot(x='popularity', y='vote_average', data=df)
plt.title("Popularity vs Vote Average")
plt.xlabel("Popularity")
plt.ylabel("Vote Average")
plt.show()

print("\nObservation:")
print("There is a slight positive relationship between popularity and ratings.")

# -----------------------------------

# GRAPH 3: Vote Count Distribution
plt.figure()
sns.histplot(df['vote_count'], bins=20)
plt.title("Distribution of Vote Count")
plt.xlabel("Vote Count")
plt.ylabel("Frequency")
plt.show()

print("\nObservation:")
print("Most shows have low vote counts, with a few having very high counts.")

# -----------------------------------

# GRAPH 4: Correlation Heatmap
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

print("\nObservation:")
print("Vote count and popularity show a stronger relationship compared to ratings.")

# -------- 6. FINAL CONCLUSION --------
print("\nFINAL CONCLUSION:")
print("""
- The dataset shows that most TV shows have moderate to high ratings.
- Popularity and vote count are somewhat related.
- Highly popular shows tend to receive more votes.
- Ratings are relatively consistent across shows.
""")
```
