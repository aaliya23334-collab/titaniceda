```
# ============================================
# TITANIC DATASET - EXPLORATORY DATA ANALYSIS
# ============================================

# -------- 1. IMPORT LIBRARIES --------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------- 2. LOAD DATASET --------
df = pd.read_csv("train (1).csv")

print("First 5 rows of dataset:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())


# -------- 3. DATA CLEANING --------

# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin column (too many missing values)
df.drop('Cabin', axis=1, inplace=True)

print("\nAfter Cleaning - Missing Values:")
print(df.isnull().sum())


# -------- 4. DATA PREPROCESSING --------

# Convert categorical to numeric
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'C':0, 'Q':1, 'S':2})


# -------- 5. DATA VISUALIZATION --------

# Graph 1: Survival Count
plt.figure()
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()

print("\nObservation 1:")
print("More passengers died than survived.")


# Graph 2: Gender vs Survival
plt.figure()
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Gender vs Survival")
plt.show()

print("\nObservation 2:")
print("Female passengers had a higher survival rate compared to males.")


# Graph 3: Age Distribution
plt.figure()
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()

print("\nObservation 3:")
print("Most passengers were between 20 and 40 years old.")


# Graph 4: Passenger Class vs Survival
plt.figure()
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Passenger Class vs Survival")
plt.show()

print("\nObservation 4:")
print("Passengers in 1st class had a higher survival rate.")


# -------- 6. FINAL CONCLUSION --------

print("\nFINAL CONCLUSION:")
print("1. Female passengers were more likely to survive.")
print("2. Higher class passengers had better survival chances.")
print("3. Age had some influence but was not the strongest factor.")
print("4. Social and economic status played a major role in survival.")
```
